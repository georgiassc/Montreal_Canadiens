# 2. How did the Montreal Canadiens perform over the years?

```python
montreal_perform = nhl[nhl['team'].str.contains('Montreal Canadiens')]
montreal_perform.plot(kind='scatter', x='rank', y='year', color='#AF1E2D', edgecolor='#192168')
plt.title('Montreal Canadiens perform over the years')
```

#  3. What was the Montreal Canadiens' best playoff season?

```python
dataset_montreal_best = nhl[nhl['team'].str.contains('Montreal Canadiens')& (nhl['rank'] == 1)][['year', 'wins']]
print ( 'The best playoff season for Montreal Canadiens was the 1993 season')
dataset_montreal_best
```

# 4. What was the Montreal Canadiens' worst playoff season?
